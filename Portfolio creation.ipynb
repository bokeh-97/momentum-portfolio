```python
all_stock_data = pd.DataFrame(columns=['Stock','Price', 'M1_return','M3_return','M6_return','M12_return','Next_1M_return'])


for stock in stock_list:
    print('\n',stock)
    try:
        df_stock = web.DataReader(stock, 'yahoo', start='2020-06-01', end='2021-08-31')
        df_stock = df_stock[['Adj Close']]
    except :
        print('no data found')
        continue
    print('df_stock.shape',df_stock.shape)
    print('null values- ',df_stock['Adj Close'].isna().sum())
    Price = df_stock.loc['2021-07-30'][0]
    new_df = pd.DataFrame()
    new_df['Monthly returns_r'] = df_stock.asfreq('BM')['Adj Close'].pct_change()#'BM' - Business month end
    new_df['1+r'] = new_df['Monthly returns_r']+1
    M1_return = new_df.iloc[-2:,-1:].values[0] - 1
    M3_return = 1
    for i in range(0,3):
        M3_return = M3_return*new_df.iloc[-(i+2):,-1:].values[0]
    M3_return = M3_return-1
    M6_return = 1
    for i in range(0,6):
        M6_return = M6_return*new_df.iloc[-(i+2):,-1:].values[0]
    M6_return = M6_return-1
    M12_return = 1
    for i in range(0,12):
        M12_return = M12_return*new_df.iloc[-(i+2):,-1:].values[0]
    M12_return = M12_return-1
    Next_1M_return = new_df.iloc[-1:,-1:].values[0]
    Next_1M_return = Next_1M_return -1
    all_stock_data = all_stock_data.append({'Stock':stock, 'Price':Price,'M1_return':M1_return[0],'M3_return':M3_return[0],'M6_return':M6_return[0],'M12_return':M12_return[0],'Next_1M_return':Next_1M_return[0]},ignore_index=True)
```

```python
all_stock_data
```

```python
freq_list = [1,3,6,12]
for row in all_stock_data.index:
    for freq in freq_list:
        all_stock_data.loc[row, f'M{freq}_return_Percentile'] = stats.percentileofscore(all_stock_data[f'M{freq}_return'], all_stock_data.loc[row, f'M{freq}_return'])/100
```

```python
all_stock_data['Momentum_factor'] = all_stock_data[['M1_return_Percentile','M3_return_Percentile','M6_return_Percentile','M12_return_Percentile']].mean(axis=1)
```

```python
test_df['Weight'] = 1/total_stocks_for_portfolio
available_capital = 250000
test_df['Number of shares'] =np.floor((available_capital*test_df['Weight'])/test_df['Price'])
```

```python
test_df['Capital Allocated'] = test_df['Number of shares']*test_df['Price']
```
